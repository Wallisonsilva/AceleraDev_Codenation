## Questão 1

Qual fração da variância consegue ser explicada pelo primeiro componente principal de `fifa`? Responda como um único float (entre 0 e 1) arredondado para três casas decimais.

```python
pca = PCA().fit(fifa)
def q1():
    pca.explained_variance_ratio_[0]
    return float(pca.explained_variance_ratio_[0].round(3))
```

## Questão 2

Quantos componentes principais precisamos para explicar 95% da variância total? Responda como un único escalar inteiro.

```python
def q2():
    cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
    component_number = np.argmax(cumulative_variance_ratio >= 0.95) + 1
    return int(component_number)
```

```python
# Plotar gráfico e verificar o número de componentes principais
evr = pca.explained_variance_ratio_
chart = sns.lineplot(np.arange(len(evr)), np.cumsum(evr))
chart.axes.axhline(0.95, ls="--", color="red")
chart.axes.axvline(q2(), ls="--", color="red")

plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance');
```

## Questão 4

Realiza RFE com estimador de regressão linear para selecionar cinco variáveis, eliminando uma a uma. Quais são as variáveis selecionadas? Responda como uma lista de nomes de variáveis.

```python
def q4():
    X = fifa.drop('Overall', axis=1)
    y = fifa['Overall']
    reg = LinearRegression().fit(X, y)
    rfe = RFE(reg, n_features_to_select=5).fit(X, y)
    return X.loc[:,rfe.get_support()].columns.tolist()
```
